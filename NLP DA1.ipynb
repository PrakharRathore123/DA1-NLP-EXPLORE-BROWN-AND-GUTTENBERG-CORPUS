```python
from nltk.corpus import gutenberg
for fileid in gutenberg.fileids():
        num_chars = len(gutenberg.raw(fileid)) 
        num_words = len(gutenberg.words(fileid))
        num_sents = len(gutenberg.sents(fileid))
        num_vocab = len(set(w.lower() for w in gutenberg.words(fileid)))
        print("Number of characters : " , round(num_chars/num_words), "Number of words: ", round(num_words/num_sents), "Number of sentences: ", round(num_words/num_vocab), " Title: " , fileid)
```

```python
from nltk.corpus import brown
# getting the categories
brown.categories()
```

```python
# getting the frequency distribution of the modals
news_text = brown.words(categories='news')
fdist = nltk.FreqDist(w.lower() for w in news_text)
modals = ['can','could','may','might','must','will']
for m in modals:
     print(m + ':', fdist[m], end=' ')
```

```python
# getting the fileids and the years
from nltk.corpus import inaugural
inaugural.fileids()
```

```python
# slicing the names and getting the years
[fileid[:4] for fileid in inaugural.fileids()]
```

```python
# generating a plot for seeing the usage of words america and citizen over the years
cfd = nltk.ConditionalFreqDist(
    (target, fileid[:4])
    for fileid in inaugural.fileids()
    for w in inaugural.words(fileid)
    for target in ['america','citizen']
    if w.lower().startswith(target)
)
```

```python
cfd.plot()
```
